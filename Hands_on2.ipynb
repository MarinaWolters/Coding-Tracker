# 3 Approximate Match

Here is the setup for *Jaccard Join*, which is really two things:

1. Take each field that is to be joined and create a set of n-grams (aka q-grams).
1. Compute the Cartesian product of all items, and assign a score based on Jaccard similarity (|set intersection| / |set union|).

````
# Approximate string matching, see 
import py_stringsimjoin as ssj
import py_stringmatching as sm


# We are going to match the strings approximately, via "n-grams" or "q-grams" (sequences of n or q characters)
# Here it's five-grams

tok = sm.QgramTokenizer(qval=3,return_set=True)
````

Find a good threshold here that seems to produce reasonable results.

````
# Now let's do a similarity join

output_pairs = ssj.jaccard_join(bn_books_df[['ISBN_13','Title','Author','Paperback']].reset_index(), am_books_df[['ISBN_13','Title','Author','Price']].reset_index(), \
                                'index', 'index', 'Title', 'Title', tok, 0.1, 
                                l_out_attrs=['Title','Author','Paperback'], r_out_attrs=['Title','Author','Price'])


display(output_pairs)

output_pairs.loc[output_pairs['_sim_score'] < 1.0] 
````
