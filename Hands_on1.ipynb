## Pandas Serie and DataFrame

**Pandas Series**

From ndarray or scalar values

```
series1 = pd.Series(data=1.,
                   index=['a', 'b', 'c', 'd', 'e'],
                   name='series_from_ndarray')
print(series1)
```

From dictionary

```
data = {'a': 1.0, 'b': 2.0, 'c': 3.0, 'd': 4.0}
series2 = pd.Series(data=data, name='series_from_dict')
print(series2*2)
```

**DataFrame**


From dictionary of Series

```
import numpy as np
data = {'column_a': pd.Series(data=np.random.randint(10, 100, 5),
                              index=['a', 'b', 'c', 'd', 'e']),
        'column_b': pd.Series(data=np.random.randint(10, 100, 5),
                              index=['a', 'b', 'c', 'd', 'e']),
        'column_c': pd.Series(data=np.random.randint(10, 100, 5),
                              index=['a', 'b', 'c', 'd', 'e'])}
df = pd.DataFrame(data=data)
df
```

From dictionary of list

```
data = {'column_a': [1, 2, 3],
        'column_b': [4, 5, 6],
        'column_c': [7, 8, 9]}
df = pd.DataFrame(data=data, index=['row_a', 'row_b', 'row_c'])
print(df)
```
